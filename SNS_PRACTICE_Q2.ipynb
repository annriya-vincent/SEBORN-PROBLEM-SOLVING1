```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

titanic=sns.load_dataset('titanic')

#fill missing values with fillana
titanic['fare']=titanic['fare'].fillna(0)
#perform minmaxscaler
scaler=MinMaxScaler()
titanic['fare_scaled']=scaler.fit_transform(titanic[['fare']])

#original fare histogram
plt.subplot(1,2,1)
sns.histplot(titanic['fare'],kde=True,color='blue',alpha=0.6)
plt.title('original fare histogram')
plt.xlabel('fare')

#scaled fare histogram
plt.subplot(1,2,2)
sns.histplot(titanic['fare_scaled'],kde=True,color='green',alpha=0.6)
plt.title('scaled fare histogram')
plt.xlabel('fare_scaled')

plt.tight_layout()
plt.show()
```

6.
a) Create a FacetGrid using the tips dataset to visualize the distribution of total_bill for different time values (Lunch/Dinner).

b) Add a stripplot on the FacetGrid, with hue for smoker.

c) Replace the stripplot with a swarmplot and compare the results.

```python
#6a
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')
tips=sns.FacetGrid(df,col='time')
tips.map(sns.scatterplot,'total_bill','tip',edgecolor='w').add_legend()
plt.show()

```

```python
#6b
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')
tips=sns.FacetGrid(df,col='time',hue='smoker')
ordd=df['total_bill'].unique()
tips.map(sns.stripplot,'total_bill','tip',edgecolor='w',jitter=True,order=ordd).add_legend()
plt.show()
```

```python
#6c
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')
tips=sns.FacetGrid(df,col='time',hue='smoker')
figure=plt.figure(figsize=(10,5))
ordd=df['total_bill'].unique()
tips.map(sns.swarmplot,'total_bill','tip',edgecolor='w').add_legend()
plt.show()
```
